```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, cauchy
```

```python
# Leggi i dati dal file
data = pd.read_csv('03.1/dataCall1Step.dat',sep=' ', header=None)

# Estrai le colonne
x = data[0]
y = data[1]
yerr = data[2]

# Crea il grafico con barre di errore
plt.figure(figsize=(10, 6))
plt.errorbar(x, y, yerr=yerr, color = '#1E88E5', ecolor='r', capsize=5)
plt.axhline(y=14.975790778311286, color='r', linestyle='--')
plt.xlabel('#throws')
plt.ylabel(r'$\langle C(S_0,0) \rangle$')
plt.title('Blocking average of call-option profit')
plt.grid(True)
plt.show()
```
